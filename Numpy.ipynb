Initial place holders in numpy


```python
# numpy of zeroes

x = np.zeros((4,5),dtype=int)
print(x)
print(type(x))
```

```python
# numpy of ones

y = np.ones((4,5) , dtype=int)
print(y)
```

```python
# numpy of AnyValue
z = np.full((2,5),5)
print(z)

z_2 = np.full((6,7),8)
print(z_2)
```

```python
#Identity matrix

iden = np.eye( 7 , dtype=int)
print(iden)

iden_1 = np.eye(4,2,dtype=int)
print(iden_1)
```

```python
# numpy array with random values , values lies between 0 - 1

a = np.random.random((4,5))
print(a)
```

```python
# numpy array with random value but with settable range

b = np.random.randint(10,100,(4,5))
print(b)
```

```python
#array of evenly spaced --> number of values requires

d = np.linspace(5,10,8,dtype=int)
print(d)
```
